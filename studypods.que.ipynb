```python
num = 7
flag = True

if num <= 1:
    flag = False
else:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            flag = False
            break

print("Prime" if flag else "Not Prime")
```

```python
num = 153
sum = 0
temp = num

while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10

if sum == num:
    print("Armstrong")
else:
    print("Not Armstrong")
```
